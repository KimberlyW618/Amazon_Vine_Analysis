```python
# Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DataFrameFunctions").getOrCreate()
```

```python
# Read in data from S3 Buckets
from pyspark import SparkFiles
#s3://kimw618-databootcamp-bucket/Resources/
url = "https://kimw618-databootcamp-bucket.s3.us-west-1.amazonaws.com/Resources/user_data.csv"
spark.sparkContext.addFile(url)
aurora = spark.read.csv(SparkFiles.get("user_data.csv"))

# Show DataFrame
aurora.show()
```
